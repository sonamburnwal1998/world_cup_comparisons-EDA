# Linear Regression with statsmodel

```python
import statsmodels.formula.api as smf
model = smf.ols("goals_z~season",data = df).fit()
```

```python
sns.regplot(x="season", y="goals_z", data=df);
```

```python
model.params
```

```python
print(model.tvalues, '\n', '\n', model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
newdata=pd.Series([5])
```

```python
data_pred=pd.DataFrame(newdata,columns=['season'])
```

```python
model.predict(data_pred)
```
